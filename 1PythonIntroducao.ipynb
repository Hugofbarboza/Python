# 2 Estruturas de decisão

```python
# IF
nota = 7

if nota >= 7:
  print('Aprovado')
  print('Parabéns')
else:
  print('Reprovado')
```

```python
# Mudando a variável
nota = 5

if nota >= 7:
  print('Aprovado')
  print('Parabéns')
else:
  print('Reprovado')
```

```python
# IF, ELSE e ELIF
nota = 6
if nota <= 4:
  print('Reprovado')
elif nota > 4 and nota <=6:
  print('Exame')
else:
  print('Aprovado')
```

# 3 Estruturas de repetição

```python
# Exemplo com while
count = 1
while count <= 5:
  print(count)
  count += 1
```

```python
# Exemplo com for
for x in range(0, 10): # o último número, 10, não entra no for
  print(x)
```

```python
for x in range(0, 10):
  print(x + 1)
```

```python
# for com incremento de 2 em 2
for x in range(0, 20, 2):
  print(x)
```

```python
# for com incremento negativo de -1 em -1
for x in range(10, 0, -1):
  print(x)
```

```python
# for com o break

for x in range(0, 10):
  if x == 4: # quando o x for 4, vai para o laço
    break
  print(x)
```

```python
# for com continue
for x in range(0, 10):
  if x == 4: # quando o x for 4, vai pular for e continuar no próximo número
    continue
  print(x)

```

# 5 Dicionários

```python
# estrutura chaver/valor
notas = {'João': 6.5,
         'Maria': 8.0,
         'Pedro': 7.7}
notas
```

```python
# tipo da variável notas
type(notas)
```

```python
# exibindo o valor de uma chave
notas['João']
```

```python
# todas as chaves com .keys
notas.keys()
```

```python
# todos os valores do dicionário com .values()
notas.values()
```

```python
# verificando a existência da chave no dicionário
print('João' in notas)
print('Fernando' in notas)
```

```python
# apagando um elemento do dicionário com del
del notas['João']
notas
```

```python
# incluindo um elemento
notas['Ana'] = 7.7
notas
```

```python
# buscando uma chave e se não for encontrada, retornar o segundo parâmetro da função .get()
notas.get('Geraldo', 'Não encontrado')
```

```python
# quando a chave existe, retorna o valor
notas.get('Maria', 'Não encontrado')
```

# 6 Sets

```python
# sets são conjuntos não ordenados de elementos não repetidos
bigdata = {'Spark', 'Hive', 'Sqoop'}
print(type(bigdata))
bigdata
```

```python
# verificando a existência de um elemento no set
'Spark' in bigdata
```

```python
# adicionando um elemento no set com .add()
bigdata.add('Hadoop')
bigdata
```

```python
# número de elementos no set
len(bigdata)
```

```python
# adicionando um elemento repetido no set
# porém o set não aceita dois elementos iguais
bigdata.add('Spark')
bigdata
```

**loc[]** -  Acesse um grupo de linhas e colunas pelo índice. É baseado principalmente em rótulos.

As entradas permitidas são:

- Linhas com apenas um índice, por ex. 5 ou 'a', (observe que 5 é interpretado como um rótulo do índice, e nunca como uma posição inteira ao longo do índice);

- Índices separados ou matriz de rótulos, por exemplo ['a', 'b', 'c'];

- Uma faixa de índices, por exemplo 'a':'f'

```python
# filtrar linhas por índice | .loc[]
df.loc[1:3] #  linhas com índice de 1 a 3 (inclusive)
```

```python
# linhas 1 e 3 | .loc[]
df.loc[[1, 3]] # índice 1 e 3
```

```python
# filtro de coluna com um determinado valor | condicional de igualdade
df.loc[df['purpose'] == 'radio/tv']
```

```python
# filtro de coluna com um determinado valor | condicional de maior que
df.loc[df['credit_amount'] > 18000]
```

```python
# atribuindo o resultado acima a uma variável, criando outro df
df2 = df.loc[df['credit_amount'] > 18000]
df2
```

```python
type(df2)
```

```python
# extraindo apenas duas colunas com  filtro condicional em outra coluna
df[['checking_status','duration']].loc[df['credit_amount'] > 18000]
```

## Renomeando coluna

## Excluindo colunas

```python
# excluindo colunas | .drop()
# parâmetro axis = 1 para remover coluna e axis = 0 para remover linha(índice)
df.drop('checking_status', axis = 1)

# outra opção, passando como parâmetro "columns"
# df.drop(columns= ['checking_status'])
```

```python
# inplace = True para fazer a exclusão definitiva da coluna
df.drop(columns= ['checking_status'], inplace = True)
```

## Dados nulos

Observação:

isna() - detecta valores ausentes e retorna um valor booleano indicando se os valores são NA. Valores NA, como None ou numpy.NaN, são mapeados para valores True. Todo o resto é mapeado para valores False. Caracteres como strings vazias '' ou numpy.inf não são considerados valores NA (a menos que você defina pandas.options.mode.use_inf_as_na = True).

isnull() - DataFrame.isnull é um pseudômino para DataFrame.isna.

```python
# verificando dados nulos | pandas.DataFrame.isnull()

df.isnull()
```

```python
# somando os dados nulos por coluna
df.isnull().sum()
```

```python
# removendo linhas onde pelo menos um campo do registro é nulo | pandas.DataFrame.dropna()
df.dropna()
```

```python
# substituindo valores nulos por parâmetro passado | pandas.DataFrame.fillna()
df.fillna(0, inplace = True) # se houver valor nulo, será substituído por 0
```

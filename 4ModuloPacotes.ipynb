#Módulos e Pacotes


Módulos são conjuntos de funcionalidades organizadas em arquivos.

Para importar um módulo, basta escrever import e o nome do módulo desejado.

Ex: **import** *nome_modulo*

```python
# importanto o módulo statistics
import statistics
```

```python
# definindo uma lista para executar algumas funções do módulo acima
z = [10,20,30,40]
```

Para usar uma função que existe dentro do módulo, é necessário escrever o nome do módulo e qualificá-lo com o nome da função.

Ex: *nome_modulo.funcao*()

```python
# função que calcula a média | statistics.mean()
statistics.mean(z)
```

```python
# função de calcular a mediana | statistics.median()
statistics.median(z)
```

Podemos também importar apenas uma função específica e não o módulo inteiro, através do comando abaixo

Ex: **from** *nome_modulo* **import** *nome_funcao*

```python
# importando funções específicas do módulo statistics | mean e median
from statistics import mean, median
```

```python
# quando fazemos a importação específica da função, não precisamos especificar o módulo antes dela
mean(z)
```

```python
median(z)
```

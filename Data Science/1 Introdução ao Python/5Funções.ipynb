#Funções

- Blocos de código reutilizáveis;
- Podem ser chamados de qualquer parte do programa;
- Podem ser chamados de outros programas.

Sintaxe para declarar uma função:



**def** (para começar a redigir a função) **nome_funcao**:

  *script da função*

Ex:


```
def nome_funcao:
  #todo o script tem que ser identado
  script_funcao
  retorno_funcao
```


Execução da função:

`nome_funcao()`

##Função sem parâmetro

````python
# Função sem parâmetro, executa uma função print
def imprime():
  print('Este é o print da função imprime')
````

````python
# Executando a função "imprime"
imprime()
````

## Função com parâmetro

````python
# Função com parâmetro n
def imprime(n):
  print(n) # a função vai pegar o que foi passado no parâmetro e inserir no script
````

````python
imprime('Este texto é o parâmetro passado para a função imprime')
````

## Função com retorno

````python
# Função que retorna o objeto passado no script dentro dela
def soma(x, y):
  adicao = x + y
  return adicao
````

````python
# Executando a função soma
soma(5,10)
````

##Função com valor default

````python
# Função com valor default (padrão)
def padrao(a=1, b=11):
  for i in range(a, b):
    print(i)
````

````python
# Executando a função padrao sem passar parâmetros
# os parâmetros default vão ser executados
padrao()
````

````python
# Executando a função padrao passando parâmetros
padrao(0,6)
````

# Funções padrão

Funções padrão são funções nativas do Python, ou seja, não precisamos importar nenhum módulo para executá-las.

https://www.w3schools.com/python/python_ref_functions.asp

Alguns exemplos:

- abs(): retorna o valor absotulo;
- max(): retorna o valor máximo;
- min(): retorna o valor mínimo;
- round(): arredonda para x casas decimais;
- sum(): executa soma.

````python
abs(-200)
````

````python
lst = [1,2,3,4,5,6]
````

````python
# maior valor da lista
max(lst)
````

````python
# menor valor da lista
min(lst)
````

````python
# arredondamento de número decimal
round(2.365842,2)
````

````python
# soma da lista
sum(lst)
````

## Modulos “Nativos”
São módulos que, para usá-los, é necessário fazer sua importação, mas não sua instalação, como no exemplo abaixo:

• import statistics


````python
import statistics
````

````python
lst = [15, 17, -5, 14, 69, 1, 17]
````

````python
# média
statistics.mean(lst)
````

````python
# mediana
statistics.median(lst)
````

````python
# moda
statistics.mode(lst)
````

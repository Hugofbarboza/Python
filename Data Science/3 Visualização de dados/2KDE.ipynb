# Kernel Density Estimate (KDE)

Um gráfico de estimativa de densidade de núcleo (KDE) é um método para visualizar a distribuição de observações em um conjunto de dados, análogo a um histograma. O KDE representa os dados usando uma curva de densidade de probabilidade contínua em uma ou mais dimensões.

A abordagem é explicada mais detalhadamente no guia do usuário.

Em relação a um histograma, o KDE pode produzir um gráfico menos confuso e mais interpretável, especialmente ao desenhar distribuições múltiplas. Mas tem o potencial de introduzir distorções se a distribuição subjacente for limitada ou não suave. Tal como num histograma, a qualidade da representação também depende da seleção de bons parâmetros de suavização.

```python
sns.kdeplot(data.iloc[:,1], color='black')\
.set(title='Distribuição de altura das árvores',
     xlabel='Altura', ylabel='Frequência');
```

Ativando a visualização do KDE no gráfico de histograma através do parâmetro KDE = True

```python
sns.histplot(data.iloc[:,1], kde=True, color='black')\
.set(title='Distribuição de altura das árvores',
     xlabel='Altura', ylabel='Frequência');
```

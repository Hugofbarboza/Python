#Gráfico de Dispersão

Os gráficos de dispersão ou Scatter plot são representações gráficas do relacionamento entre duas variáveis numéricas. O Scatter plot utiliza pontos para representar essa relação, cada ponto representa o valor de uma variável no eixo horizontal e o valor de outra variável no eixo vertical.

```python
# Comparando as variáveis Girth(Circunferência) e Volume | Gráfico de dispersão Matplotlib.pyplot
plt.scatter(x=data.Girth, y=data.Volume, color='grey', marker='*')
plt.title('Circunferência x Volume')
plt.xlabel('Circunferência')
plt.ylabel('Volume');
```

```python
# Comparando as variáveis Girth(Circunferência) e Volume | Gráfico de linha Matplotlib.pyplot
plt.plot(data.Girth, data.Volume)
plt.title('Circunferência x Volume')
plt.xlabel('Circunferência')
plt.ylabel('Volume');
```

```python
# Comparando as variáveis Girth(Circunferência) e Volume | Gráfico de dispersão Seaborn
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
sns.regplot(x = data.Girth, y = data.Volume, data = data,
            x_jitter = 0.3,
            fit_reg = False).set(title='Circunferência x Volume',#fit_reg=False para tirar a linha de tendência
                                 xlabel='Circunferência',
                                 ylabel='Volume');
```

x_jitter - adiciona uma aleatoridade aos dados, para não se sobreporem, os pontos vão se mover um pouco de forma aleatória

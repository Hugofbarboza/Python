## Distribuição binomial

A Distribuição Binomial é uma distribuição de probabilidade discreta, ou seja: serve para calcular a probabilidade da ocorrência de eventos discretos (eventos discretos são eventos que podemos contar). Diferente da distribuição normal que calcula a probabilidade da ocorrência de eventos contínuos, podendo ser mensurados em intervalos.

Pré-requisitos para que possamos trabalhar com a distribuição binomial
- Número fixo de experimentos;
- Cada experimento pode ter 2 resultados apenas: sucesso ou fracasso;
- A probabilidade de sucesso deve ser a mesma em cada experimento;
- Os experimentos são independentes.

```python
# Importando a função binom da biblioteca scipy
from scipy.stats import binom
```

```python
# Ao jogar uma moeda 5 vezes, qual a probabilidade
# da mesma cair com a face "cara" 3 vezes

# binom.pmf(eventos, experimentos, probabilidades)
prob = binom.pmf(3, 5, 0.5)
prob
```

```python
# Ao passar por 4 sinais de 4 tempos, qual a probabilidade de pegar sinal verde?
# nenhuma(0), 1, 2, 3 ou 4 vezes seguidas?
binom.pmf(0, 4, 0.25) + binom.pmf(1, 4, 0.25) + binom.pmf(2, 4, 0.25) + binom.pmf(3, 4, 0.25) + binom.pmf(4, 4, 0.25)
```

```python
# nenhuma
binom.pmf(0, 4, 0.25)
```

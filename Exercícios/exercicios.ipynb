```python
# primeira forma
cateto_a = float(input('Digite o cateto adjacente do triângulo: '))
cateto_o = float(input('Digite o cateto oposto do triângulo: '))

hip = hypot(cateto_a, cateto_o)

print(f'A hipotenusa do triângulo é {round(hip, 2)}')
```

#24ª Exercício
O Índice de Massa Corporal (IMC) é utilizado para mensurar o peso ideal de uma pessoa. Escreva um programa que peça o nome, a idade , o peso e a altura do usuário. Ao final calcule e mostre o resultado do seu IMC e classifique este resultado de acordo com a regra a seguir.

IMC<17 - Muito abaixo do peso ideal

17<=IMC<18,5 - Abaixo do peso

18,5<=IMC<25 - Peso normal

25<=IMC<30 - Acima do peso

30<=IMC<35 - Obesidade I

35<=IMC<40 - Obesidade II (severa)

IMC>=40 - Obesidade III (mórbida)

Lembre que: IMC=massa/(altura*altura)

Fonte: https://pt.wikipedia.org/wiki/%C3%8Dndice_de_massa_corporal

```python
import time
```

```python
print('Cálculo do IMC \n  \n***********')

nome = input('Digite seu nome: ').title().strip()
idade = int(input('Digite sua idade: '))
peso = float(input('Digite o seu peso em quilos: '))
altura = float(input('Digite sua altura em metros: '))


imc = round(peso/(altura**2),2)
print('\n \n')

print('Processando seus dados')
for i in range(6):
  print('*')
  time.sleep(1)

if imc < 17:
  print(f'{nome}, seu IMC é {imc}, muito abaixo do peso ideal')
elif imc >=17 and imc < 18.5:
  print(f'{nome}, seu IMC é {imc}, abaixo do peso')
elif imc >= 18.5 and imc < 25:
  print(f'{nome}, seu IMC é {imc}, peso normal')
elif imc >= 25 and imc < 30:
  print(f'{nome}, seu IMC é {imc}, acima do peso')
elif imc >= 30 and imc < 35:
  print(f'{nome}, seu IMC é {imc}, obesidade I')
elif imc >= 35 and imc < 40:
  print(f'{nome}, seu IMC é {imc}, obesidade II (severa)')
else:
  print(f'{nome}, seu IMC é {imc}, obesidade III (mórbida)')
```

Escreva um programa que receba dois números de ponto flutuante e mostre na tela o maior número digitado. Considere a possibilidade de o usuário digitar dois números iguais.

```python
import time
```

```python
print('COMPARAÇÃO DE NÚMEROS \n')
print('Digite dois números para compararmos qual é o maior \n')

a = float(input('Escreva o primeiro número: '))
b = float(input('Escreva o segundo número: '))

print('\n Comparando')
for i in range(4):
  print('.')
  time.sleep(1)

if a == b:
  print('Os números são iguais')
elif a > b:
  print(f'O número {a} é maior que o número {b}')
else:
  print(f'O número {b} é maior que o número {a}')
```

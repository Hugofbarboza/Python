## Gráfico de linha

```python
# plotando o gráfico
plt.title('Linha do gráfico')
plt.plot(x, y)
plt.xlabel('Variável do eixo x')
plt.ylabel('Variável do eixo y');
```

## Gráfico de barras

```python
# difinindo uma tupla com os nomes das linguagens de programação
linguagens = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
linguagens
```

```python
# criando um intervalo com a quantidade de items da tupla linguagens
x = np.arange(len(linguagens))
x
```

```python
# atribuindo valores a um índice de performance fictício para as linguagens de programação
y_performance = [10,8,6,4,2,1]
y_performance
```

```python
# criando script para plotar o gráfico de barras
plt.title('Uso de linguagens de programação')
plt.bar(x, y_performance)
plt.xticks(x, linguagens)
plt.xlabel('Linguagens de programação')
plt.ylabel('Taxa de uso');
```

## Gráfico de pizza (setores)

Demonstrando a porcentagem de vendas de celulares a partir do seu sistema operacional

```python
# definindo os rótulos (sistemas operacionais)
s_operacional = ['Android', 'Blackberry', 'Windows Phone', 'iOS']
s_operacional
```

```python
# definindo uma quantidade de vendas fictícia
vendas = [40,20,10,30]
vendas
```

```python
# definindo as cores das "fatias" do gráfico
cores = ['lightyellow', 'lightgreen', 'lightcoral', 'lightskyblue']
cores
```

```python
# distância das fatias com a finalidade de destaque
explode = (0.1,0,0,0)
```

```python
# plotando o gráfico de pizza
plt.pie(vendas,
        explode=explode, # distância das fatias
        labels=s_operacional, # rótulos das fatias
        colors=cores,
        autopct = '%.f%%', # porcentagem dentro das fatias
        shadow = True,  # efeito de sombra
        startangle = 0);
```

## Histograma

```python
# criando uma lista com números aleatórios para plotar no gráfico de histograma
dados = [43,45,53,56,56,57,58,66,67,73,74,79, 80,80,81,81,81,82,83,83,84,88,89,91,
91,92,92,97,99,99,100,100,101,102,102,102,103,104,107,108,109,113,114,118,121,123,126,128,
137,138,139,144,145,147,156,162,174,178,179,184,191,198,211,214,243,249,329,380,403,511,522,598]
```

```python
# definindo intervalos de frequência
intervalos = [0,100,200,300,400,500,600]
```

```python
# plotando o gráfico de histograma
# grades
plt.grid(color = '#cccccc', linestyle='-')

# histograma
plt.hist(dados, bins=intervalos, zorder=2)
plt.title('Gráfico de histograma')
plt.xlabel('Intervalos')
plt.ylabel('Frequência');
```

```python
plt.grid(color = '#cccccc', linestyle='-')
plt.hist(dados, bins=intervalos, zorder=2)
```

```python
# mostrando os valores no histograma com o método text()

plt.grid(color='#cccccc', linestyle='-', zorder=1)
aux = plt.hist(dados, zorder=2, bins=intervalos)

for i in np.arange(len(intervalos)-1):
    plt.text(aux[1][i]+30, aux[0][i], str(aux[0][i]))
```

## Gráfico de dispersão

Relação entre peso e altura de dois grupos fictícios de pessoas

```python
# Grupo A
peso1 = [57,58.2,58.6,59.6,59.8,60.2,60.5,60.6,60.7,61.3,61.3,61.4,61.8,61.9,62.3]
altura1 = [100.7,195.6,94.3,127.1,111.7,159.7,135,149.9,124.3,112.9,176.7,110.2,123.9,161.9,107.8]
```

```python
# Grupo B
peso2 = [62.9,63,63.1,63.2,63.3,63.4,63.4,63.4,63.5,63.6,63.7,64.1,64.3,64.3,64.7,64.8,65]
altura2 = [151.8,156.3,136.1,124.2,156.3,130,181.2,255.9,163.1,123.1,119.5,179.9,114.8,174.1,108.8,105.4,141.4]
```

```python
# Plotando o gráfico com plt.scatter(eixo_x, eixo_y)

plt.scatter(peso1, altura1, marker='*', color='r', label='GrupoA')
plt.scatter(peso2, altura2, marker='.', color='b', label='GrupoB')
plt.title('Altura x Peso', fontsize=20)
plt.xlabel('Peso - Kg', fontsize=16)
plt.ylabel('Altura - cm', fontsize=16)
plt.legend();
```

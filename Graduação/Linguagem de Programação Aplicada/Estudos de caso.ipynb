# Estudo de caso 1

Apresentação de um gráfico de barras para demonstrar o ganho de medalhas de cinco países. Os dados serão criados manualmente (fictícios) apenas para demonstrar o uso de um gráfico de barras segmentadas.

```python
# Importando as bibliotecas necessárias

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando uma lista com países de amostra
paises = ['EUA', 'China', 'Brasil', 'Argentina', 'Alemanha']
```

```python
# Dados referentes a cada país da amostra e suas respectivas medalhas de
# ouro, prata e bronze; dados fictícios

bronze = np.array([38, 17, 26, 19, 15])
prata = np.array([37, 23, 18, 18, 10])
ouro = np.array([46, 27, 26, 19, 17])
```

```python
# Criando um array com índices para o eixo x do gráfico
ind = np.arange(len(paises))
ind
```

```python
# Plotando os dados acima em um gráfico e barras
# Título do gráfico
plt.title('Medalhas - Olimpíadas 2018')

# Plots
plt.bar(ind, ouro, label='Ouro', color='gold', bottom=prata+bronze)
plt.bar(ind, prata, label='Prata', color='silver', bottom=bronze)
plt.bar(ind, bronze, label='Bronze', color='#CD853F')

# Ativando a caixa de legendas no gráfico
plt.legend()

# Rótulos das colunas
plt.xticks(ind, paises)

# Descrição dos eixos x e y
plt.xlabel('Países')
plt.ylabel('Quantidade de medalhas');
```

# Estudo de caso 2

Um setor público precisa visualizar os resultados das eleições presidenciais de 2018. É necessário criar um gráfico para demonstrar a quantidade de votos dos três candidatos mais votados. Tal gráfico deve evidenciar as porcentagens de cada um dos candidatos nas regiões Sudeste, Nordeste, Sul, Norte e Centro-Oeste

```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo uma lista com os nomes das regiões de interesse
regioes = ['SUDESTE', 'NORDESTE', 'SUL', 'NORTE', 'CENTRO-OESTE']
regioes
```

```python
# Definindo os votos dos candidatos arrays seguindo a ordem respectiva das regiões acima
bolsonaro = np.array([53.965, 28.187, 58.446, 50.357, 57.677])
haddad = np.array([20.74, 49.824, 19.213, 30.331, 20.59])
ciro = np.array([11.937, 16.214, 8.786, 7.657, 9.707])
```

```python
# Definindo um índice numérico atrelado a quantidade de regiões
index = np.arange(len(regioes))
index
```

```python
# Plotando os gráficos acima em um gráfico de barras
# Título
plt.title('Distribuição de presidenciáveis nas regiões brasileiras',
          pad=10)

# Plot
plt.bar(index, ciro, label='Ciro', color='yellow')
plt.bar(index, haddad, label='Haddad', color='blue', bottom=ciro)
plt.bar(index, bolsonaro, label='Bolsonaro', color='red', bottom=haddad+ciro)

# Nomes das colunas
plt.xticks(index, regioes)

# Labels do eixo x e y
plt.xlabel('Regiões')
plt.ylabel('Quantidade de votos')

# Ativando a legenda com localização no campo superior esquerdo
plt.legend(fontsize=6, loc='upper left');
```

```python
media_horas = []
rotulos = []
```

# Estudo de caso 4

1 - Calcule e mostre – no prompt – a porcentagem de alunos com as notas – grade – nas faixas < 60, = 70, = 80, > 90.

2 - Apresente os dados graficamente para as notas – grade – nas faixas <= 60, > 60 e <= 70, > 70 e <= 80, > 80, destacando a maior taxa.

3 - Apresente a maior e menor notas para ambos os gêneros. Para cada valor encontrado,
imprima o nome e sobrenome do primeiro aluno da lista.

4 - Crie um gráfico de dispersão horas/nota.

```python
# Importando bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando arquivo excel para o estudo do caso
dataset = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Cruzeiro do Sul/Linguagem de Programação Aplicada/gradedata.xlsx')
dataset.head()
```

```python
dataset.info()
```

##1

```python
# 1 - Calcule e mostre – no prompt – a porcentagem de alunos
# com as notas – grade – nas faixas < 60, = 70, = 80, > 90.

porcentagem_notas = dataset['grade'].value_counts(normalize=True) * 100
```

```python
porcentagem_notas = pd.DataFrame(porcentagem_notas)
porcentagem_notas
```

```python
# Notas menores que 60
porcentagem_notas[porcentagem_notas.index < 60].sum()
```

```python
# Notas iguais a 70
porcentagem_notas[porcentagem_notas.index == 70]
```

```python
# Notas iguais a 80
porcentagem_notas[porcentagem_notas.index == 80]
```

```python
porcentagem_notas[porcentagem_notas.index > 90].sum()
```

##2

```python
# 2 - Apresente os dados graficamente para as notas – grade – nas faixas
# <= 60, > 60 e <= 70, > 70 e <= 80, > 80, destacando a maior taxa.

# <= 60
menorIgual60 = len(dataset[dataset['grade'] <= 60]) * 100 / len(dataset)
menorIgual60
```

```python
# > 60 e <= 70

maior60Menor70 = len(dataset[(dataset['grade'] > 60) & (dataset['grade'] <= 70)]) * 100 / len(dataset)
maior60Menor70
```

```python
# > 70 e <= 80

maior70Menor80 = len(dataset[(dataset['grade'] > 70) & (dataset['grade'] <= 80)]) * 100 / len(dataset)
maior70Menor80
```

```python
# > 80

maior80 = len(dataset[dataset['grade'] > 80]) * 100 / len(dataset)
maior80
```

```python
# Passando os intervalos para serem plotados no gráfico
intervalos = ('< 60', '> 60 e <= 70', '> 70 e <=80', '>80')
```

```python
# Selecionando uma valores obtidos acima para inserir no gráfico
valores = [menorIgual60, maior60Menor70, maior70Menor80, maior80]
valores
```

```python
# Definindo cores aleatórias para as 'fatias'
colors = ['blue', 'red', 'grey', 'yellow']
```

```python
# medidas e espaçamento entre as 'fatias'
explode = (0, 0, 0, 0.2)
```

```python
# Plotando o gráfico
plt.pie(valores, labels=intervalos,
        explode=explode,
        colors=colors, autopct='%1.1f%%')

plt.legend(intervalos, fontsize=7, loc='upper left')
plt.axis('equal');
```

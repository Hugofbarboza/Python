Exemplo

# Método

Exemplo: classe Círculo, que armazena valores de raio e retorna a área do círculo

```python
class Circulo():
    PI = 3.14 # atributo de classe

    def __init__(self, initRaio):
        self.raio = initRaio # atributo de instância

    # método de acesso para configurar um novo raio
    def setRaio(self, newRaio):
        self.raio = newRaio

    # método de acesso para retornar o valor atual do raio
    def getRaio(self):
        return self.raio

    # método específico para calcular a área do círculo
    def calcArea(self):
        return Circulo.PI * (self.raio ** 2)
```

```python
# instanciando um objeto
circulo1 = Circulo(5)
```

```python
# pegando o valor do raio definido acima
circulo1.getRaio()
```

```python
# acessando o valor da constante PI
circulo1.PI
```

```python
# alterando o valor do raio de 5 para 10
circulo1.setRaio(10)
circulo1.getRaio()
```

```python
# Calculando o valor da área do círculo
circulo1.calcArea()
```

```python
# Acessando métodos especiais na classe Circulo
dir(Circulo)
```

```python
# Definindo novamente a classe Circulo com o método especial __str__

class Circulo():
    PI = 3.14 # atributo de classe

    def __init__(self, initRaio):
        self.raio = initRaio # atributo de instância

    # método de acesso para configurar um novo raio
    def setRaio(self, newRaio):
        self.raio = newRaio

    # método de acesso para retornar o valor atual do raio
    def getRaio(self):
        return self.raio

    # método específico para calcular a área do círculo
    def calcArea(self):
        return Circulo.PI * (self.raio ** 2)

    #Método __str__
    def __str__(self):
        return (f'Círculo de raio: {self.raio}, área: {self.calcArea()}')
```

```python
circulo2 = Circulo(7)
```

# Herança


```python
# classe FormaGeometrica

class FormaGeometrica():
    def __init__(self, initX, initY):
        self.x = initX
        self.y = initY

    def getX(self):
        return self.x

    def setX(self, newX):
        self.x = newX

    def getY(self):
        return self.y

    def setY(self, newY):
        self.y = newY
```

```python
# classe Circulo

class Circulo(FormaGeometrica):
    PI = 3.14

    def __init__(self, initRaio, initX, initY):
        # chamar o __init__ da classe pai
        super().__init__(initX, initY)
        self.raio = initRaio

    def setRaio(self, newRaio):
        self.raio = newRaio

    def getRaio(self):
        return self.raio

    def calcArea(self):
        return Circulo.PI * (self.raio ** 2)

    def __str__(self):
        return (f'Círculo de raio: {self.raio}, área do raio: {self.calcArea()}')
```

```python
# classe Quadrado

class Quadrado(FormaGeometrica):
    def __init__(self, initLado, initX,initY):
        super().__init__(initX, initY)
        self.lado = initLado

    def setLado(self, newLado):
        self.lado = newLado

    def getLado(self):
        return self.lado

    def calcArea(self):
        return (self.lado ** 2)

    def __str__(self):
        return (f'Quadrado de lado: {self.lado}, área: {self.calcArea()}')
```

```python
forma1 = Quadrado(5, 1, 6)
```

```python
print(forma1)
```

```python
forma1.getX(), forma1.getY()
```

```python
forma1.calcArea()
```

 # Encapsulamento de dados

Atributo e método "fracamente privados"

```python
# classe Conta com 2 atributos e 1 método "privados"

class Conta():
    def __init__(self, initNum, initSaldo):
        self._numero = initNum
        self._saldo = initSaldo

    def _adicionarBonus(self):
        self._saldo += 1000
```

```python
conta1 = Conta(1, 1200)
```

```python
conta1.numero
```

```python
conta1._numero
```

```python
conta1._adicionarBonus()
```

```python
conta1._saldo
```

Atributos e métodos "fortemente privados"

```python
class Conta():
    def __init__(self, initNum, initSaldo):
        self.__numero = initNum
        self.__saldo = initSaldo

    def __adicionarBonus(self):
        self.__saldo += 1000

    def getSaldo(self):
        return self.__saldo
```

```python
conta1 = Conta(1, 2000)
```

```python
conta1.numero
```

```python
conta1._Conta__saldo
```

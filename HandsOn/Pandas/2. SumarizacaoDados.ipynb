# Bibliotecas e DataFrame

```python
# importando bibliotecas pandas e numpy
import pandas as pd
import numpy as np
```

- `no.random.seed()`: Fixa números aleatórios;
- `np.arange()`: Retorna valores uniformemente espaçados dentro de um determinado intervalo;
- `np.random.randint()`: Retorna números inteiros aleatórios sendo o primeiro parâmetro inclusivo, o segundo exclusivo e o terceiro a quantidade de números;
- `np.random.choice()`: Gera uma amostra aleatória de uma determinada matriz. Primeiro parâmetro é a amostra aleatória e o segundo parâmetro é a quantidade de valores.

```python
np.random.seed(10)
# quantidade de observações
n = 100

escolaridade = ['Ensino Fundamental', 'Ensino Médio', 'Ensino Superior']

cidades = ['São Paulo', 'Recife', 'Salvador', 'Salvador', 'Rio de Janeiro']

sexo = ['Masculino', 'Feminino']

# dicionário com dados
data = {
    'Id': np.arange(1, n+1),
    'Idade': np.random.randint(18, 70, size = n),
    'Renda': np.random.randint(1_500, 10_000, size = n),
    'Sexo': np.random.choice(sexo, size = n),
    'Escolaridade': np.random.choice(escolaridade, size = n),
    'Cidade': np.random.choice(cidades, size = n)
}

df = pd.DataFrame(data)
```

##pivot_table - girando a tabela

```python
desemprego_data.head()
```

```python
# definindo um novo DataFrame com o índice sendo a coluna Data
# as colunas a coluna state e os valores a coluna UnemploymentRate
desemprego_total = desemprego_data.pivot_table(
    index = 'Date',
    columns = 'state',
    values = 'UnemploymentRate'
)
```

```python
desemprego_total.head()
```

```python
# filtra os dados para estados selecionados
states = [
    "Arizona", "California", "Florida", "Illinois",
    "Michigan", "New York", "Texas"
]
desemprego = desemprego_total[states]
```

##agg - Aplique uma função ou um nome de função a um dos eixos do DataFrame

```python
def high_or_low(data):

    if data.mean() < 6.5:
        return 'Low'
    return 'High'
```

```python
desemprego.agg(high_or_low)
```

Função de agregação escrita de outro modo:

```python
def high_or_low(data):

    if data.mean() < 6.5:
        result = 'Low'
        return result
    else:
        result = 'High'
        return result
```

```python
desemprego.agg(high_or_low, axis = 1).head(20)
```

Podemos aplicar mais de uma métrica de agregação por vez:

```python
desemprego.agg([min, high_or_low,max])
```

```python
desemprego.agg([min,high_or_low,max],axis = 1).head()
```
